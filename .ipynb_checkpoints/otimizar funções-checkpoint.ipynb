## otimização  
 $\min   \ \ x_1 x_4  \left(x_1 + x_2 + x_3\right) + x_3$  
 $\mathrm{s.t.} \ \ \ \ \quad x_1 x_2 x_3 x_4 \ge 25$  
 $x_1^2 + x_2^2 + x_3^2 + x_4^2 = 40$  
 $x_0 = (1,5,5,1)$

```python
def objective(x):
    x1 = x[0]
    x2 = x[1]
    x3 = x[2]
    x4 = x[3]
    return(x1*x4*(x1+x2+x3)+x3)
def constraint1(x):
    return x[0]*x[1]*x[2]*x[3] - 25
def constraint2(x):
    return x[0]**2 + x[1]**2 + x[2]**2 +x[3]**2 - 40
```

```python
x0 = [1,5,5,1]
objective(x0)
```

```python
bound_range =[1,5]
bounds = (bound_range,bound_range,bound_range,bound_range)
con1 = {'type': 'ineq', 'fun': constraint1}
con2 = {'type': 'eq', 'fun': constraint2}
cons = [con1,con2]
```

```python
sol = optimize.minimize(objective, x0, method="SLSQP", bounds=bounds,
                        constraints= cons)
```

```python
print('objective function = ',sol.fun)
print('solution values = ',sol.x)
```
